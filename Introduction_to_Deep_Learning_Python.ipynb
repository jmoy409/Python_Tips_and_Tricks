## Back Propagation
- technique used to calculate slopes you need to optimize more complex deep learning models

#### Forward Propagation sends input data through the hidden layers and into the output layer, Back Propagation takes the error from the output layer, propagates it through the hidden layers to the weights of the input layer.
- allows gradient descent to update all weights in neural network (by getting gradients for all weights)

### Back Propagation Process
- Trying to estimate the slope of the loss function w.r.t. each weight in our network.

### Process:
- Do forward propagation to make a prediction and calculate the error before we do back propagation
- Go back one layer at a time, use a formula for slopes, every weight feeds from some input node into some output node

### The 3 things we need to multiply to get the slope for that weight are:
- 1) node value feeding into that weight (done in forward propagation)
- 2) slope of loss function w.r.t. node it feeds in (done in backward propagation)
- 3) slope of activation function at the node it feeds into (if relu, negative is 0, postive is input itself. so if postive, slope would be 1)

### Slope of node values are the sum of the slopes for all weights that come out of them

## The relationship between forward and backward propagation:
- If you have gone through 4 iterations of calculating slopes (using backward propagation) and then updated weights, how many times must you have done forward propagation? 4

## Thinking about backward propagation
- If your predictions were all exactly right, and your errors were all exactly 0, the slope of the loss function with respect to your predictions would also be 0. In that circumstance, which of the following statements would be correct?
- The updates to all weights in the network would also be 0.

### Understanding model optimization

#### why optimization is hard
- optimal value for any one weight depends on the values of other weights
- even if slopes tell us to update weights x and weights y, may not improve our model meaningfully
- updates too small (if learning rate is low) or too large (if learning rate is high)
- the easiest way to see the effect of different learning rates is to use the simpliest optimizer: schocastic gradient descent (SGD)
- can specify learning rate with LR argument
- use SGD with a different learning rate
- even if LR is well tuned, can still run into dying-neuron problem

### Dying Neuron Problem
- neuron takes a value less than 0 for all rows of your data
- ex) once a node starts getting negative inputs, it may continue to getting negative inputs 

### Vanishing Gradients
- occurs when many layers have very small slopes (flat part of tanh curve)
- slopes get close to 0 which meant updates in backprop were close to 0

- Start with 1 hidden layer and a good amount of nodes per layer, measure the mean squared error
- Increase the nodes per layer, measure MSE again, if it drops, good! Lets try adding another hidden layer
- Add another hidden layer with good amount of nodes per layer, if it decreases MSE, good!
- If adding another hidden layer or nodes per layer doesnt decrease MSE, STOP
